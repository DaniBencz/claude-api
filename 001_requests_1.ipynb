```python
# Create API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
# Make request
message = client.messages.create(
  model=model,
  max_tokens=1000,
  messages=[
    {
      "role": "user",
      "content": "What is quantum computing? Answer in one sentence"
    }
  ]
)
```
